```python
def path(dir):
    pdf_files = []
    for file in os.listdir(dir):
        if file.endswith(".pdf"):
            pdf_files.append(os.path.join(dir, file))
    return pdf_files
pdf_files = path(r"C:\Users\karu0\Machine Learning\Contracts\CUAD_v1\CUAD_v1\full_contract_pdf\Part_I\Hosting")

try:
    l = []
    success  = 0
    n = 0
    for i in pdf_files:
        if i != 0:
            l.append("END OF CONTRACT")
        doc = fitz.open(i)
        success += 1
        for page in doc:
            output = page.get_text("blocks")
            prev_blockid = 0
            for block in output:
                if  block[6] == 0:                        
                    l.append(block[4])
                    
except e:
    n += 1
    print(e)
print(success)
print(n)
print(len(pdf_files))
```

```python
df = df.dropna(ignore_index=True)
```

```python
from sklearn import preprocessing 

label_encoder = preprocessing.LabelEncoder() 
df['Class_labels']= label_encoder.fit_transform(df['Class']) 

len(df['Class_labels'].unique())

dict(zip(label_encoder.classes_, label_encoder.transform(label_encoder.classes_)))
```

```python
from bertopic import BERTopic
from bertopic.vectorizers import ClassTfidfTransformer
from bertopic.dimensionality import BaseDimensionalityReduction
from sklearn.linear_model import LogisticRegression

docs = df['Text']
y = df['Class_labels']


empty_dimensionality_model = BaseDimensionalityReduction()
clf = LogisticRegression()
ctfidf_model = ClassTfidfTransformer(reduce_frequent_words=True)

topic_model= BERTopic(
        umap_model=empty_dimensionality_model,
        hdbscan_model=clf,
        ctfidf_model=ctfidf_model
)
topics, probs = topic_model.fit_transform(docs, y=y)
```
